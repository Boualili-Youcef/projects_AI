```python
train, test = racine.split_train_test(data)
train_accuracies = []
test_accuracies = []

depth = 30
for max_depth in range(depth):
    racine = da.Noeud(profondeur_max=max_depth)
    racine.grow(train)
    train_accuracies.append(racine.precision(train))
    
    test_accuracies.append(racine.precision(test))

print(train_accuracies)
print(test_accuracies)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(range(1,depth+1), train_accuracies, label="Prcision Apprentissage", color="blue", marker="o")
plt.plot(range(1,depth+1), test_accuracies, label="Prcision Test", color="red", marker="x")
plt.xlabel("Profondeur maximale")
plt.ylabel("Prcision (%)")
plt.xticks(ticks=[x for x in range(1, len(train_accuracies) + 1, 1)]) 
plt.legend()
plt.grid()
plt.show()
```

```python
# Initialisation des listes pour stocker les prcisions
train_accuracies = []
test_accuracies = []

# Tester pour diffrentes valeurs de alpha
for alpha in np.linspace(0, 1,20):  # Essayer diffrentes valeurs de alpha
    racine.grow(train)  # Construire l'arbre sur l'ensemble d'entranement
    print("Prcision avant lagage (Train):", racine.precision(train))
    # Appliquer l'lagage
    racine.elagage(alpha)
    print("Prcision avant lagage (Test):", racine.precision(test))
    # Calculer la prcision sur l'ensemble d'entranement
    train_accuracy = racine.precision(train)
    train_accuracies.append(train_accuracy)
    
    # Calculer la prcision sur l'ensemble de test
    test_accuracy = racine.precision(test)
    test_accuracies.append(test_accuracy)
# Avant lagage



# Appliquer l'lagage
racine.elagage(alpha)

# Aprs lagage
print("Prcision aprs lagage (Train):", racine.precision(train))
print("Prcision aprs lagage (Test):", racine.precision(test))

# Vrification du nombre d'lments dans les listes
print("Longueur de train_accuracies:", len(train_accuracies))
print("Longueur de test_accuracies:", len(test_accuracies))

# Tracer les rsultats
plt.figure(figsize=(10, 6))
plt.plot(np.linspace(0, 1,20), train_accuracies, label="Prcision Apprentissage", color="blue", marker="o")
plt.plot(np.linspace(0, 1,20), test_accuracies, label="Prcision Test", color="red", marker="x")
plt.xlabel("Alpha ()")
plt.ylabel("Prcision (%)")
plt.legend()
plt.grid(True)
plt.title("volution de la prcision en fonction de alpha")
plt.show()
```
